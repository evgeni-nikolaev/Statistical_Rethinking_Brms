## 添加预测变量
\begin{aligned} \log \frac{\operatorname{Pr}\left(y_{i} \leq k\right)}{1-\operatorname{Pr}\left(y_{i} \leq k\right)} &=\alpha_{k}-\phi_{i} \\ \phi_{i} &=\beta x_{i} \end{aligned}
**要在每个截距的基础上减去φ(即βx)而不是加的原因是，β的取值为正数意味着预测变量x增加导致结果的期望增加**

```r
# first, we needed to specify the `logistic()` function, which is apart of the `dordlogit()` function
logistic <- function(x) {
    p <- 1 / (1 + exp(-x))
    p <- ifelse(x == Inf, 1, p)
    p
    }

# now we get down to it
dordlogit <- 
  function(x, phi, a, log = FALSE) {
    a  <- c(as.numeric(a), Inf)
    p  <- logistic(a[x] - phi)
    na <- c(-Inf, a)
    np <- logistic(na[x] - phi)
    p  <- p - np
    if (log == TRUE) p <- log(p)
    p
    }

(pk <- dordlogit(1:7, 0, fixef(b11.1)[, 1]))
```

### 零膨胀泊松分布函数

```r
dzip <- function(x, p, lambda, log = TRUE) {
    ll <- ifelse(
        x == 0,
        p + (1 - p) * exp(-lambda),
        (1 - p) * dpois(x, lambda, log = FALSE)
    )
    if (log == TRUE) ll <- log(ll)
    return(ll)
}

p_b11.4      <- posterior_summary(b11.4)[2, 1]
lambda_b11.4 <- posterior_summary(b11.4)[1, 1] %>% exp()

tibble(x = 0:4) %>% 
  mutate(density = dzip(x = x, 
                        p = p_b11.4, 
                        lambda = lambda_b11.4, 
                        log = F)) %>% 
  
  ggplot(aes(x = x, y = density)) +
  geom_col(fill = canva_pal("Green fields")(4)[4]) +
  xlab("Manuscripts completed") +
  theme_hc() +
  theme(plot.background = element_rect(fill = "grey92"))
```

# 过度离散结果

## Beta二项
Beta分布有两个参数，平均概率$\bar{p}$和尺度参数θ。尺度参数θ刻画了分布的分散情况。当θ=2时，概率取值(从0到1)是等可能分布的。当θ>2时，概率分布变得更加集中。当 0<2时，分布分散的情况更加严重，接近0和1的区域的概率密度高于均值。
$$\operatorname{Beta}(y | \alpha, \beta)=\frac{y^{\alpha-1}(1-y)^{\beta-1}}{\mathrm{B}(\alpha, \beta)}$$
- 期望：$\mu=\frac{\alpha}{\alpha+\beta}$
- 分散度：$\kappa=\alpha+\beta$
- 标准差：$\sigma=\sqrt{\mu(1-\mu) /(\kappa+1)}$  

### 参数定义

```r
pbar  <- 0.5
theta <- 5

ggplot(data = tibble(x = seq(from = 0, to = 1, by = .01))) +
  geom_ribbon(aes(x    = x, 
                  ymin = 0, 
                  ymax = rethinking::dbeta2(x, pbar, theta)),
              fill = canva_pal("Green fields")(4)[1]) +
  scale_x_continuous("probability space", breaks = c(0, .5, 1)) +
  scale_y_continuous(NULL, breaks = NULL) +
  ggtitle(expression(paste("The ", beta, " distribution")),
          subtitle = expression(paste("Defined in terms of ", mu, " (i.e., pbar) and ", kappa, " (i.e., theta)"))) +
  theme_hc() +
  theme(plot.background = element_rect(fill = "grey92"))
```

```r
betaABfromMeanKappa <- function(mean, kappa) {
  if (mean <= 0 | mean >= 1) stop("must have 0 < mean < 1")
  if (kappa <= 0) stop("kappa must be > 0")
  a <- mean * kappa
  b <- (1.0 - mean) * kappa
  return(list(a = a, b = b))
}

betaABfromMeanKappa(mean = pbar, kappa = theta)
```

### 建模
\begin{aligned} A_{i} & \sim \text { BetaBinomial }\left(n_{i}, \bar{p}_{i}, \theta\right) \\ \operatorname{logit}\left(\bar{p}_{i}\right) &=\alpha \\ \alpha & \sim \operatorname{Normal}(0,10) \\ \theta & \sim \operatorname{HalfCauchy}(0,1) \end{aligned}
其中结果变量 A 是录取结果 admit, n 是提交的申请数目 applications，这个模型没有用任何预测变量，但是如果你要添加预测变量，可以将它们加到第二行的线性模型中。

```r
library(rethinking)
data(UCBadmit)
d <- UCBadmit
rm(UCBadmit)
detach(package:rethinking, unload = T)
library(brms)
d
```

```r
beta_binomial2 <- 
  custom_family(
    "beta_binomial2", dpars = c("mu", "phi"),
    links = c("logit", "log"), lb = c(NA, 0),
    type = "int", vars = "trials[n]"
  )

stan_funs <- "
  real beta_binomial2_lpmf(int y, real mu, real phi, int T) {
    return beta_binomial_lpmf(y | T, mu * phi, (1 - mu) * phi);
  }
  int beta_binomial2_rng(real mu, real phi, int T) {
    return beta_binomial_rng(T, mu * phi, (1 - mu) * phi);
  }
"

stanvars <- 
  stanvar(scode = stan_funs, block = "functions")

b11.5 <-
  brm(data = d, 
      family = beta_binomial2,                           # 自定义似然
      admit | trials(applications) ~ 1,
      prior = c(prior(normal(0, 2), class = Intercept),
                prior(exponential(1), class = phi)),
      iter = 4000, warmup = 1000, cores = 4, chains = 4,
      stanvars = stanvars,  # note our `stanvars`
      seed = 11)

print(b11.5)

system("osascript -e \'display notification \"程序\" with title \"跑完了\"\'")
beepr::beep(sound = 4)
```

```r
post <- posterior_samples(b11.5)

head(post)
```

```r
# 中位数与基于分位数的 95% 置信区间

post %>% 
  tidybayes::median_qi(inv_logit_scaled(b_Intercept)) %>% 
  mutate_if(is.double, round, digits = 3)

# 中位数是 0.4, 但相应的分位数区间非常宽
```

### Beta分布后验均值

```r
library(tidyverse)

set.seed(11)

lines <-
  post %>% 
  mutate(iter  = 1:n(),                               # 排绘制贝塔分布的后验均值
         p_bar = inv_logit_scaled(b_Intercept)) %>%   
  sample_n(size = 100) %>%                            # 从后验分布中抽取 100 个贝塔分布
  expand(nesting(iter, p_bar, phi),
         x = seq(from = 0, to = 1, by = .005)) %>% 
  mutate(density = pmap_dbl(list(x, p_bar, phi), rethinking::dbeta2))

str(lines)
```

```r
lines %>% 
  ggplot(aes(x = x, y = density)) + 
  stat_function(fun = rethinking::dbeta2,
                args = list(prob  = mean(inv_logit_scaled(post[, 1])),
                            theta = mean(post[, 2])),
                color = canva_pal("Green fields")(4)[4],
                size = 1.5) +
  geom_line(aes(group = iter),
            alpha = .2, color = canva_pal("Green fields")(4)[4]) +
  scale_y_continuous(limits = c(0, 3), breaks = NULL) +
  xlab("probability admit") +
  theme_hc() +
  theme(plot.background = element_rect(fill = "grey92"))
```

模型 b11.5 中 100 个从后验分布中抽取的 p 和 θ 的取值组合。蓝色粗线是后验样本均值对应的贝塔分布。细线代表 100 个从后验分布中抽取的 p 和 θ 取值组合对应的贝塔分布。平均的情况倾向于低录取率，小于 0.5。但最可能的分布照顾到了录取率很高的院系

```r
expose_functions(b11.5, vectorize = TRUE)

# required to use `predict()`
log_lik_beta_binomial2 <- 
    function(i, draws) {
        mu  <- draws$dpars$mu[, i]
        phi <- draws$dpars$phi
        N   <- draws$data$trials[i]
        y   <- draws$data$Y[i]
        beta_binomial2_lpmf(y, mu, phi, N)
  }


posterior_predict_beta_binomial2 <- 
    function(i, draws, ...) {
        mu  <- draws$dpars$mu[, i]
        phi <- draws$dpars$phi
        N   <- draws$data$trials[i]
        beta_binomial2_rng(mu, phi, N)
  }

pp_expect_beta_binomial2 <-
    function(draws) {
        mu <- draws$dpars$mu
        trials <- draws$data$trials
        trials <- matrix(trials, nrow = nrow(mu), ncol = ncol(mu), byrow = TRUE)
        mu * trials
}

predict(b11.5) %>%
  as_tibble() %>% 
  transmute(ll = Q2.5,
            ul = Q97.5) %>%
  # the fitted intervals
  bind_cols(
    fitted(b11.5) %>%
  as_tibble()
  ) %>% 
  # Tte original data used to fit the model
  bind_cols(b11.5$data) %>% 
  mutate(case = 1:12) %>% 
  
  # plot
  ggplot(aes(x = case)) +
  geom_linerange(aes(ymin = ll / applications, 
                     ymax = ul / applications),
                 color = canva_pal("Green fields")(4)[1], 
                 size = 2.5, alpha = 1/4) +
  geom_pointrange(aes(ymin = Q2.5  / applications, 
                      ymax = Q97.5 / applications, 
                      y = Estimate/applications),
                  color = canva_pal("Green fields")(4)[4],
                  size = 1/2, shape = 1) +
  geom_point(aes(y = admit/applications),
             color = canva_pal("Green fields")(4)[2],
             size = 2) +
  scale_x_continuous(breaks = 1:12) +
  scale_y_continuous(limits = c(0:1), breaks = c(0, .5, 1)) +
  labs(subtitle = "Posterior validation check",
       y        = "Admittance probability") +
  theme_hc() +
  theme(plot.background = element_rect(fill = "grey92"),
        axis.ticks.x    = element_blank(),
        legend.position = "none")

system("osascript -e \'display notification \"程序\" with title \"跑完了\"\'")
beepr::beep(sound = 4)
```

核査模型 b11.5 的后验分布。对每个`case`预测录取率，横轴表示不同的`case`，黑点表示数据中观测到的录取率，空心圆是后验均值$\bar{p}$以及 95% 的分位数区间，+号表示 95% 的录取率预测分位数区间。这里不同院系对应的录取率不一样。模型并不知道有院系变量，因为我们并没有告诉模型这一点。但是它确实检测出了不同行之间的异方差性，并且通过贝塔分布估计预测异方差的情况。但这些区间如此之大，几乎不能算是对模型的认可
