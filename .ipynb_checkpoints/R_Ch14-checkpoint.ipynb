## 结果变量误差
使用`Mean=Observation Mean`且`SD=Observation sd`的高斯分布是合乎逻辑的选择，因为如果我们斤了解误差的标准差，则该误差的最大熵分布将是高斯

```r
d %>% 
  mutate(Divorce_distribution = str_c("Divorce ~ Normal(", Divorce, ", ", Divorce.SE, ")")) %>% 
  select(Loc, Divorce_distribution) %>% 
  head()
```

`brms`中有两种方法来囊括标准中的测量误差：
- `se()`语法，公式为`<response> | se(<se_response>, sigma = TRUE)`。`se`代表标准差；
- `mi()`语法，公式为`<response> | mi(<se_response>)`。 遵循缺失值逻辑，得出了标准数值值的贝叶斯缺失值填补法。`mi()`语法是基于brms较新的缺失数据能力。
\begin{align*}
\text{Divorce}_{\text{estimated}, i} & \sim \text{Normal} (\mu_i, \sigma) \\
\mu     & = \alpha + \beta_1 \text A_i + \beta_2 \text R_i \\
\text{Divorce}_{\text{observed}, i}  & \sim \text{Normal} (\text{Divorce}_{\text{estimated}, i}, \text{Divorce}_{\text{standard error}, i}) \\
\alpha  & \sim \text{Normal} (0, 10) \\
\beta_1 & \sim \text{Normal} (0, 10) \\
\beta_2 & \sim \text{Normal} (0, 10) \\
\sigma  & \sim \text{HalfCauchy} (0, 2.5).
\end{align*}

```r
# put the data into a `list()`
dlist <- list(
    div_obs = d$Divorce,
    div_sd  = d$Divorce.SE,
    R       = d$Marriage,
    A       = d$MedianAgeMarriage)

# here we specify the initial (i.e., starting) values
inits      <- list(Yl = dlist$div_obs)
inits_list <- list(inits, inits)

# fit the models
b14.1_se <- 
  brm(data = dlist, family = gaussian,
      div_obs | se(div_sd, sigma = TRUE) ~ 0 + Intercept + R + A,
      prior = c(prior(normal(0, 10), class = b),
                prior(cauchy(0, 2.5), class = sigma)),
      iter = 5000, warmup = 1000, cores = 2, chains = 2,
      seed = 14,
      control = list(adapt_delta = 0.99,
                     max_treedepth = 12),
      inits = inits_list)

b14.1_mi <- 
  brm(data = dlist, family = gaussian,
      div_obs | mi(div_sd) ~ 0 + Intercept + R + A,
      prior = c(prior(normal(0, 10), class = b),
                prior(cauchy(0, 2.5), class = sigma)),
      iter = 5000, warmup = 1000, cores = 2, chains = 2,
      seed = 14,
      control = list(adapt_delta = 0.99,
                     max_treedepth = 12),
      save_mevars = TRUE,  # note this line for the `mi()` model
      inits = inits_list)

system("osascript -e \'display notification \"程序\" with title \"跑完了\"\'")
beepr::beep(sound = 4)
```

各个模型的初始值（即`inits`）不同。即使我们使用了`inits = inits_list`为两种模型都进行了编码，`fit@inits`显示两者之间也存在差异

```r
b14.1_se$fit@inits
cat("===========================================================")
b14.1_mi$fit@inits
```

```r
print(b14.1_se)
cat("===========================================================")
print(b14.1_mi)
```

```r
library(broom)

tidy(b14.1_se) %>%
  mutate_if(is.numeric, round, digits = 2)
cat("===========================================================")
tidy(b14.1_mi) %>%
  mutate_if(is.numeric, round, digits = 2)
```

基于`mi()`语法的模型b`14.1_mi`包含了所有50个评价值的后验摘要。`se()`只提供了整体数据。`rethinking`包将这50个值按`div_est[i]`格式索引；`brms::mi()`则将其索引为`Yl[i]`--大差不差。因此，虽然这两种`brms`方法都考虑了测量误差，但`mi()`方法与`rethinking`的理念更一致。

`brm`公式中的`0 + intercept`部分将截距保留在未变换的数据中，导致了类似于`rethinking`的结果。当先验较为模糊的时候(《Statistical Rethinking》中的许多模型只使用弱正则先验)问题不大。但这个模型对截距语法相当敏感。我对应用数据分析的一般建议是这样的:**如果你的预测变量不是做的均值中心化，那么在使用`brms:::brm()`时，默认为`0+Intercept`语法的公式参数。否则你设定的先验可能会翻车。**

## 结果变量与预测的误差
`me()`语句来指定包含误差的预测变量，其形式为`me(predor, sd_predictor)`，其中`sd_predictor`是数据中的一个向量，表示测量误差的大小，假定为标准差度量

```r
# the data
dlist <- list(
  div_obs = d$Divorce,
  div_sd  = d$Divorce.SE,
  mar_obs = d$Marriage,
  mar_sd  = d$Marriage.SE,
  A       = d$MedianAgeMarriage)

# the `inits`
inits      <- list(Yl = dlist$div_obs)
inits_list <- list(inits, inits)

# the models
b14.2_se <-
  brm(data = dlist, family = gaussian,
      div_obs | se(div_sd, sigma = TRUE) ~ 0 + Intercept + me(mar_obs, mar_sd) + A,
      prior = c(prior(normal(0, 10), class = b),
                prior(cauchy(0, 2.5), class = sigma)),
      iter = 5000, warmup = 1000, chains = 3, cores = 3,
      seed = 14,
      control = list(adapt_delta = 0.95),
      save_mevars = TRUE) # note the lack if `inits`

b14.2_mi <- 
  brm(data = dlist, family = gaussian,
      div_obs | mi(div_sd) ~ 0 + Intercept + me(mar_obs, mar_sd) + A,
      prior = c(prior(normal(0, 10), class = b),
                prior(cauchy(0, 2.5), class = sigma)),
      iter = 5000, warmup = 1000, cores = 2, chains = 2,
      seed = 14,
      control = list(adapt_delta = 0.99,
                     max_treedepth = 12),
      save_mevars = TRUE,
      inits = inits_list)

print(b14.2_se)
cat("===========================================================")
print(b14.2_mi)
```

```r
tidy(b14.2_se) %>%
  mutate_if(is.numeric, round, digits = 2)
cat("===========================================================")
tidy(b14.2_mi) %>%
  mutate_if(is.numeric, round, digits = 2)
```

在`brm()`函数中指定`save_mevars = TRUE`，以便保存通过使用`me()`参数获得的误差调整变量的后验样本。如果不这样做，像`predict()`这样的函数可能会翻车

```r
data_error <-
  fitted(b14.2_mi) %>%
  as_tibble() %>%
  bind_cols(d)

color <- viridis_pal(option = "C")(7)[3]

p3 =  ggplot(data_error, aes(x = Divorce.SE, y = Estimate - Divorce)) +
  geom_hline(yintercept = 0, linetype = 2, color = "white") +
  geom_point(alpha = 2/3, size = 2, color = color)

color_y <- viridis_pal(option = "C")(7)[7]
color_p <- viridis_pal(option = "C")(7)[2]

p4 = posterior_samples(b14.2_mi) %>%
  select(starts_with("Xme")) %>%
  gather() %>%
  # this extracts the numerals from the otherwise cumbersome names in `key` and saves them as integers
  mutate(key = str_extract(key, "\\d+") %>% as.integer()) %>%
  group_by(key) %>%
  summarise(mean = mean(value)) %>%
  bind_cols(data_error) %>%
  
  ggplot(aes(x = mean, y = Estimate)) +
  geom_segment(aes(xend = Marriage, yend = Divorce),
               color = "white", size = 1/4) +
  geom_point(size = 2, alpha = 2/3, color = color_y) +
  geom_point(aes(x = Marriage, y = Divorce), 
             size = 2, alpha = 2/3, color = color_p) +
  scale_y_continuous(breaks = seq(from = 4, to = 14, by = 2)) +
  labs(x = "Marriage rate (posterior)" , y = "Divorce rate (posterior)") +
  coord_cartesian(ylim = c(4, 14.5))

options(repr.plot.width=18, repr.plot.height=9)
cowplot::plot_grid(p3, p4, ncol = 2)
```

黄点表示模型收缩后的结果。紫色的是原始数据。事实证明，我们的`brms`模型比McElreath的模型更积极地进行正则化

# 缺失值

## 填补新皮质数据

```r
library(rethinking)
data(milk)
d <- milk

d <-
  d %>%
  mutate(neocortex.prop = neocortex.perc / 100,
         logmass        = log(mass))
detach(package:rethinking, unload = T)
library(brms)
rm(milk)

# prep data
data_list <- 
  list(
    kcal      = d$kcal.per.g,
    neocortex = d$neocortex.prop,
    logmass   = d$logmass)

data_list
```

\begin{align*}
\text{kcal}_i           & \sim \text{Normal} (\mu_i, \sigma) \\
\mu_i                   & = \alpha + \beta_1 \text{neocortex}_i + \beta_2 \text{logmass}_i \\
\text{neocortex}_i      & \sim \text{Normal} (\nu, \sigma_\text{neocortex}) \\
\alpha                  & \sim \text{Normal} (0, 100) \\
\beta_1                 & \sim \text{Normal} (0, 10) \\
\beta_2                 & \sim \text{Normal} (0, 10) \\
\sigma                  & \sim \text{HalfCauchy} (0, 1) \\
\nu                     & \sim \text{Normal} (0.5, 1) \\
\sigma_\text{neocortex} & \sim \text{HalfCauchy} (0, 1).
\end{align*}

```r
b_model <- 
  #主要的 `kcal` 模型，mi(neocortex) 表明在建模时存在含缺失值变量
  bf(kcal ~ 1 + mi(neocortex) + logmass) + 
  # 对`neocortex`缺失值建模，符合某一个分布
  bf(neocortex | mi() ~ 1) + 
  # 两个模型的残差不想管
  set_rescor(FALSE)

get_prior(data = data_list, 
          family = gaussian,
          b_model)
```

```r
b14.3 <- 
  brm(data = data_list, 
      family = gaussian,
      b_model,  # here we insert the model
      prior = c(prior(normal(0, 100), class = Intercept, resp = kcal),
                prior(normal(0.5, 1), class = Intercept, resp = neocortex),
                prior(normal(0, 10),  class = b),
                prior(cauchy(0, 1),   class = sigma,     resp = kcal),
                prior(cauchy(0, 1),   class = sigma,     resp = neocortex)),
      iter = 1e4, warmup = 2000, chains = 2, cores = 2,
      seed = 14)


# 删除带有NA的案例的模型
b14.3cc <- 
  brm(data = data_list, 
      family = gaussian,
      kcal ~ 1 + neocortex + logmass,
      prior = c(prior(normal(0, 100), class = Intercept),
                prior(normal(0, 10), class = b),
                prior(cauchy(0, 1), class = sigma)),
      iter = 1e4, warmup = 2000, chains = 2, cores = 2,
      seed = 14)

tidy(b14.3) %>%
  mutate_if(is.numeric, round, digits = 2)

tidy(b14.3cc) %>%
  mutate_if(is.numeric, round, digits = 2)
```

- 展示了 12 个缺失观测的填补，后半部分是线性回归的参数估计
- 通过加入观测不完全的样本，新皮层比例的后验均值从 2.77 变成 1.91, 体重的均值从 -0.1 变成-0.07。因此通过使用所有的样本，推断的关系变强了

```r
nd <-
  tibble(neocortex = seq(from = .5, to = .85, length.out = 30),
         logmass   = median(data_list$logmass))

f_b14.3 <-
  fitted(b14.3, newdata = nd) %>%
  as_tibble() %>%
  bind_cols(nd)

f_b14.3 %>%
  glimpse()
```

```r
# 加入新皮质估计值
f_b14.3_mi <-
  tidy(b14.3) %>%
  filter(str_detect(term, "Ymi")) %>%
  bind_cols(data_list %>%
              as_tibble() %>%
              filter(is.na(neocortex))
            )

f_b14.3_mi %>% head()
```

## 改善插值模型
\begin{array}{l}\text { neocortex }_{i} \sim \operatorname{Normal}\left(\nu_{i}, \sigma_{\text {neocortex }}\right) \\ \qquad \begin{aligned} \nu_{i} &=\alpha_{\text {neocortex }}+\gamma_{1} \log \operatorname{mass}_{i} \end{aligned} \\ \qquad \begin{aligned} \alpha_{\text {neocortex }} & \sim \operatorname{Normal}(0.5,1) \\ \gamma_{1} & \sim \operatorname{Normal}(0,10) \end{aligned}\end{array}

```r
# define the model
b_model <- 
  bf(kcal ~ 1 + mi(neocortex) + logmass) + 
  bf(neocortex | mi() ~ 1 + logmass) + # 为`neocortex`添加`logmass`预测变量
  set_rescor(FALSE)

# fit the model
b14.4 <- 
  brm(data = data_list, 
      family = gaussian,
      b_model,
      prior = c(prior(normal(0, 100), class = Intercept, resp = kcal),
                prior(normal(0.5, 1), class = Intercept, resp = neocortex),
                prior(normal(0, 10),  class = b),
                prior(cauchy(0, 1),   class = sigma,     resp = kcal),
                prior(cauchy(0, 1),   class = sigma,     resp = neocortex)),
      iter = 1e4, chains = 2, cores = 2,
      seed = 14)

tidy(b14.4) %>%
  mutate_if(is.numeric, round, digits = 2)

system("osascript -e \'display notification \"程序\" with title \"跑完了\"\'")
beepr::beep(sound = 4)
```

```r
f_b14.4 <-
  fitted(b14.4, newdata = nd) %>%
  as_tibble() %>%
  bind_cols(nd)

f_b14.4_mi <-
  tidy(b14.4) %>%
  filter(str_detect(term, "Ymi")) %>%
  bind_cols(data_list %>%
              as_tibble() %>%
              filter(is.na(neocortex))
            )

f_b14.4 %>%
  glimpse()
cat("===================================================")
f_b14.4_mi %>%
  glimpse()
```

# 元分析
元分析是贝叶斯多层模型的特例
> 关于元分析的讨论有时并不精确，特别是当研究的主要重点是检验任何一项研究都不存在影响的假说时。我们的重点是估计有意义的参数，为了这个目标，在接受研究在某种广义上具有可比性的前提下，似乎有三种可能性。第一种可能性是，我们将这些研究视为彼此完全相同的复制，也就是说，我们将所有研究中的个体看作是来自一个共同人口的独立样本，具有相同的结果衡量标准等等。第二种可能性是，这些研究之间存在着巨大的差异，以至于任何一项研究的结果都无法提供其他研究结果的信息。第三种，也是更普遍的可能性是，我们认为这些研究是可以交换的，但不一定是完全相同或完全不相关的；换句话说，我们允许不同的研究之间存在差异，但这些差异不会先验地预期会对一个研究产生有利于另一个研究的可预测效应...........这第三种可能性代表了这两个极端之间的连续体，而正是这种可交换的模型（具有未知的超参数特征的人口分布）构成了我们贝叶斯分析的基础..........

>元分析的第一个潜在的估计值，或者说一般的分层结构问题，是效应大小分布的平均数，因为这代表了所有研究的总体 "平均 "效应，可被视为可与观察到的研究交换。其他可能的估计变量是任何一项观察到的研究中的效应大小和另一项可比较（可交换）的未观察到的研究中的效应大小。(第125-126页，原文强调)

基本的贝叶斯分析遵循：$y_{i} \sim \operatorname{Normal}\left(\theta_{i}, \sigma_{i}\right)$，其中，$y_i$ = 单一研究的效应大小的点估计值，i假定从$Mean = \theta_i$正态分布中取得。元分析中的数据通常是个别研究的统计摘要。从本章中得到的一个明显的教训是，这些估计本身就有误差，而这些误差应该在元分析模型中充分表达出来。来自研究 i 的标准误被指定为$\sigma_i$，而正态分布的标准差是由点估计得出的。请注意，我们在这里不是在估计$\sigma_i$。这些值我们直接从原始研究中提取。  
在模型的基础上，我们进一步假设研究i本身只是从相关的研究群体中抽取一个样本，每个样本都有自己的效应大小。因此，我们假定$\theta_i$本身具有如下形式的分布：$\theta_{i} \sim \operatorname{Normal}(\mu, \tau)$，其中μ是元分析效应（即群体均值），τ是围绕该均值的变异，也可以认为是$\sigma_\tau$

案例：
>在世界范围内，大多数儿童（80%）被父母打屁股或以其他方式体罚（儿基会，2014年）。在美国和世界各地，父母是否应该打孩子的屁股来纠正孩子的不当行为，这个问题在美国和世界各地都存在着伦理、宗教和人权等方面的争论（Gershoff，2013年）。关于父母使用打屁股或体罚与孩子的行为、情绪、认知和身体结果之间的关联，已经进行了几百项研究，使打屁股成为研究最多的育儿方式之一。从这数百项研究中，我们学到了什么？

```r
rm(list = ls())
spank <- readxl::read_excel("spank.xlsx")
glimpse(spank)
```

书中的效应值是通过`Cohen'd`(两组间的均值差的标准化系数)定义的：
$$d=\frac{\mu_{\text {treatment }}-\mu_{\text {comparison }}}{\sigma_{\text {pooled }}}, \sigma_{\text {pooled }}=\sqrt{\frac{\left(\left(n_{1}-1\right) \sigma_{1}^{2}\right)+\left(\left(n_{2}-1\right) \sigma_{2}^{2}\right)}{n_{1}+n_{2}-2}}$$

```r
spank <-
  spank %>% 
  mutate(se = (ul - ll) / 3.92)   # 置信区间的标准化

glimpse(spank)
```

在心理学中，很少看到`Cohen'd`绝对值大于±1的，所以在没有更具体的领域知识的情况下（我没有），Normal(0,1)似乎是一个合理的起点。就像McElreath用HalfCauchy(0, 1)作为组水平标准差的默认先验值一样，在这里用它作为我们的元分析τ参数也是合理的
$$
\begin{array}{l}\mathrm{d}_{i} \sim \operatorname{Normal}\left(\theta_{i}, \sigma_{i}=\mathrm{se}_{i}\right) \\ \theta_{i} \sim \operatorname{Normal}(\mu, \tau) \\ \mu \sim \operatorname{Normal}(0,1) \\ \tau \sim \text { HalfCauchy }(0,1)\end{array}
$$
您可能会注意到的一件事是我们的`se(se)`函数不含sigma参数。如果您回想起14.1节，我们在测量误差模型中制定了`sigma = T`。`brms`中`se()`默认`sigma = FALSE`，因此，我们无法像使用研究中的原始数据进行分析那样对sigma进行估算。希望这是有道理的。每个研究项目i的d值不确定性已经被编码到se函数中。这把我们引到了另一点。我们通常对数据摘要执行荟萃分析。在我的领域中，也许在您的领域中，由于历史的偶然性，因为研究人员并不公开数据。因此，效应大小摘要是我们通常最好的。但是，时代变了。如果你的元分析的所有研究的原始数据都是可用的，你可以直接拟合一个多层次模型，其中的数据嵌套在研究中。哎呀，您甚至可以探索可变σ的分布式建模方法，并指定类似`sigma ~ 0 + study`或类似`sigma ~ 1 + (1 | study)`。

```r
b14.5 <- 
  brm(data = spank, family = gaussian,
      d | se(se) ~ 1 + (1 | study),
      prior = c(prior(normal(0, 1), class = Intercept),
                prior(cauchy(0, 1), class = sd)),
      iter = 8000, warmup = 2000, cores = 4, chains = 4,
      seed = 14)

print(b14.5)
```

总体`Cohen's d`约为0.38。我们对τ的估计值为0.26，表明我们有相当大的研究之间的变异性。你可能要问的一个问题是： 这些`Cohen's d`s到底是在衡量什么？我们已经把这个问题编码到了结果向量中

```r
spank %>% 
  distinct(outcome) %>% 
  knitr::kable()
```

```r
spank %>% 
  distinct(study) %>% 
  count()
```

首先，除了 "成人对体罚的支持 "可能是个例外，所有的结果都是负的。我们对 "儿童攻击性 "和 "成人心理健康问题 "等低值相关的条件更感兴趣。其次，从数据编码的方式来看，较大的效应被解释为与儿时受过体罚相关的更多负面行为。也就是说，我们的分析表明，对儿童体罚与更糟糕的结果有关。可能不明显的是，即使数据中有111个案例，但只有76个不同的研究。换句话说，有些研究有多个结果。为了更好地拟合研究（和结果）水平的变异，让我们拟合一个交叉分类的Bayesian元分析，让我们联想到第13章中的交叉分类黑猩猩模型

```r
b14.6 <- 
  brm(data = spank, family = gaussian,
      d | se(se) ~ 1 + (1 | study) + (1 | outcome),
      prior = c(prior(normal(0, 1), class = Intercept),
                prior(cauchy(0, 1), class = sd)),
      iter = 8000, warmup = 2000, cores = 4, chains = 4,
      seed = 14)

print(b14.6)
```

```r
# we'll want this to label the plot
label <-
  tibble(tau   = c(.12, .3),
         y     = c(15, 10),
         label = c("sigma['outcome']", "sigma['study']"))

# wrangle
posterior_samples(b14.6) %>% 
  select(starts_with("sd")) %>% 
  gather(key, tau) %>% 
  mutate(key = str_remove(key, "sd_") %>% str_remove(., "__Intercept")) %>% 
  
  # plot
  ggplot(aes(x = tau)) +
  geom_density(aes(fill = key),
               color = "transparent") +
  geom_text(data = label,
            aes(y = y, label = label, color = label),
            parse = T, size = 5) +
  scale_fill_viridis_d(NULL, option = "B", begin = .5) +
  scale_color_viridis_d(NULL, option = "B", begin = .5) +
  scale_y_continuous(NULL, breaks = NULL) +
  xlab(expression(tau)) +
  theme(panel.grid = element_blank())
```

在这一点上，最大的问题是，研究之间的变异性比结果间更多。但我还是想了解一下各个结果的情况。这里，我们将使用`tidybayes:::geom_halfeyeh()`来帮助我们制作我们版本的森林地块，并使用`tidybayes:::spread_draws()`来帮助我们进行最初的纠结。

```r
library(tidybayes)
options(repr.plot.width=12, repr.plot.height=9)

b14.6 %>% 
  spread_draws(b_Intercept, r_outcome[outcome,]) %>%
  # add the grand mean to the group-specific deviations
  mutate(mu = b_Intercept + r_outcome) %>%
  ungroup() %>%
  mutate(outcome = str_replace_all(outcome, "[.]", " ")) %>% 

  # plot
  ggplot(aes(x = mu, y = reorder(outcome, mu), fill = reorder(outcome, mu))) +
  geom_vline(xintercept = fixef(b14.6)[1, 1], color = "grey33", size = 1) +
  geom_vline(xintercept = fixef(b14.6)[1, 3:4], color = "grey33", linetype = 2) +
  geom_halfeyeh(.width = .95, size = 2/3, color = "white") +
  scale_fill_viridis_d(option = "B", begin = .2) +
  labs(x = expression(italic("Cohen's d")),
       y = NULL) +
  theme(panel.grid   = element_blank(),
        axis.ticks.y = element_blank(),
        axis.text.y  = element_text(hjust = 0))
```

背景中的实线和虚线垂直黑线标记了均值（即荟萃分析效果）及其95％区间。但是无论如何，结果之间并没有太多的变异
> 当假设可交换性时，我们假设不存在可能构成更复杂模型基础的重要协变量，这种假设（也许是误导性的）在元分析中被广泛采用。如果有其他信息（除了数据(x，y)之外，还有其他的信息来区分 J 元分析中的研究，那么可交换模型就不合适吗？在这种情况下，我们可以将模型的框架扩展到观察到的数据和协变量的可交换性，例如使用层次回归模型

荟萃分析中提到的一个重要协变量是研究类型。他们基于荟萃分析的76篇论文包含了Within和Between设计。在Spank数据中，我们用Between向量和Within向量对这些信息进行了假编码。within=1-between。

```r
spank %>% 
  count(between)
```

在我的模型中使用假指标时，我更愿意让主要群体作为参考类别。因此，我通常会用少数群体来命名变量。本研究中大多数案例是被试内设计的。因此，我们将继续把被试间变量加入到模型中。在这时，我们将练习使用 "0+Intercept"语法。

```r
b14.7 <- 
  brm(data = spank, family = gaussian,
      d | se(se) ~ 0 + intercept + between + (1 | study) + (1 | outcome),
      prior = c(prior(normal(0, 1), class = b),
                prior(cauchy(0, 1), class = sd)),
      iter = 8000, warmup = 2000, cores = 4, chains = 4,
      seed = 14)

print(b14.7)
```

```r
color <- viridis_pal(option = "B")(7)[5]

posterior_samples(b14.7) %>% 
  
  ggplot(aes(x = b_between, y = 0)) +
  geom_halfeyeh(.width = c(.5, .95), color = "white", fill = color) +
  scale_y_continuous(NULL, breaks = NULL) +
  xlab("Overall difference for between- vs within-participant designs") +
  theme(panel.grid   = element_blank())
```
