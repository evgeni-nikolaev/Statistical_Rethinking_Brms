# 创建交互项

```r
library(rethinking)
data(rugged)
d <- rugged
detach(package:rethinking, unload = T)
library(brms)
rm(rugged)
head(d)
```

```r
library(tidyverse)

# make the log version of criterion
d <- 
  d %>%
  mutate(log_gdp = log(rgdppc_2000))

# 选取有GDP的国家
dd <-
  d %>%
  filter(complete.cases(rgdppc_2000))

# 划分数据集为非洲国家与非洲外国家
d.A1 <-
  dd %>%
  filter(cont_africa == 1)

d.A0 <-
  dd %>%
  filter(cont_africa == 0)

# 前两个模型的预测log_gdp是单变量的
```

```r
b7.1 <-
  brm(data = d.A1, family = gaussian,
      log_gdp ~ 1 + rugged,       #以崎岖度作为一阶自变量建立回归
      prior = c(prior(normal(8, 100), class = Intercept),
                prior(normal(0, 1), class = b),
                prior(uniform(0, 10), class = sigma)),
      iter = 10000, warmup = 1000, chains = 4, cores = 4,
      seed = 7)

b7.2 <-
  update(b7.1, 
         newdata = d.A0)
```

```r
library(ggthemes)

nd <- 
  tibble(rugged = seq(from = 0, to = 6.3, length.out = 30))

f_b7.1 <-
  fitted(b7.1, newdata = nd) %>%
  as_tibble() %>%
  bind_cols(nd)

f_b7.2 <-
  fitted(b7.2, newdata = nd) %>%
  as_tibble() %>%
  bind_cols(nd)

# 联结两个模型数据, with `f_b7.1` stacked atop `f_b7.2`
f <-
  full_join(f_b7.1, f_b7.2) %>%
  mutate(cont_africa = rep(c("Africa", "not Africa"), each = 30))

dd %>%
  mutate(cont_africa = ifelse(cont_africa == 1, "Africa", "not Africa")) %>%
  
  ggplot(aes(x = rugged)) +
  geom_smooth(data = f,
              aes(y = Estimate, ymin = Q2.5, ymax = Q97.5,
                  fill = cont_africa, color = cont_africa),
              stat = "identity", 
              alpha = 1/4, size = 1/2) +
  geom_point(aes(y = log_gdp, color = cont_africa),
             size = 2/3) +
  scale_colour_pander() +
  scale_fill_pander() +
  scale_x_continuous("Terrain Ruggedness Index", expand = c(0, 0)) +
  ylab("log GDP from year 2000") +
  theme_pander() + 
  theme(text = element_text(family = "Times"),
        legend.position = "none") +
  facet_wrap(~cont_africa)
```

```r
b7.3 <- add_criterion(b7.3, c("loo", "waic")) #使用 brms 中的 add_criterion 函数添加信息准则
b7.4 <- add_criterion(b7.4, c("loo", "waic"))

loo_compare(b7.3, b7.4,
            criterion = "waic")

loo_compare(b7.3, b7.4,
            criterion = "loo")

model_weights(b7.3, b7.4,
              weights = "waic") %>% 
  round(digits = 3)
```

## 加入线性交互效应有效
\begin{array}{l}Y_{i} \sim \operatorname{Normal}\left(\mu_{i}, \sigma\right) \\ \mu_{i}=\alpha+\gamma_{i} R_{i}+\beta_{\mathrm{A}} A_{i} \\ \gamma_{i}=\beta_{R}+\beta_{A R} A_{i}\end{array}

```r
b7.5 <-
  update(b7.4,
         formula = log_gdp ~ 1 + rugged*cont_africa) 
```

```r
b7.5 <- add_criterion(b7.5, c("loo", "waic"))

l <- loo_compare(b7.3, b7.4, b7.5,
                 criterion = "loo")

print(l, simplify = F)
```

```r
cbind(loo_diff = l[, 1] * -2,
      se       = l[, 2] *  2)
```

```r
b7.5b <-
  update(b7.5,
         formula = log_gdp ~ 1 + rugged + cont_africa + rugged:cont_africa) 
```

# 线性交互效应的对称性

## 布里丹之驴的交互效应
- 包含交互效应的模型表达式：
\begin{array}{l}Y_{i} \sim \operatorname{Normal}\left(\mu_{i}, \sigma\right) \\ \mu_{i}=\alpha+\gamma_{i} R_{i}+\beta_{A} A_{i} \\ \gamma_{i}=\beta_{R}+\beta_{A R} A_{i}\end{array}
- 将$\gamma_i$带入$\mu_i$展开：
\begin{aligned} \mu_{i} &=\alpha+\left(\beta_{R}+\beta_{A R} A_{i}\right) R_{i}+\beta_{A} A_{i} \\ &=\alpha+\beta_{R} R_{i}+\beta_{A R} A_{i} R_{i}+\beta_{A} A_{i} \end{aligned}
- 将$A_i$的系数合并：  
$$\mu_{i}=\alpha+\beta_{R} R_{i}+\underbrace{\left(\beta_{A}+\beta_{A R} R_{i}\right)}_{G} A_{i}$$

```r
# new predictor data for `fitted()`
nd <- 
  tibble(rugged      = rep(range(dd$rugged), times = 2),
         cont_africa = rep(0:1,              each = 2))

# `fitted()`
f <-
  fitted(b7.5, newdata = nd) %>%
  as_tibble() %>%
  bind_cols(nd) %>% 
  mutate(ox = rep(c(-0.05, 0.05), times = 2))

# augment the `dd` data a bit
dd %>% 
  mutate(ox          = ifelse(rugged > median(rugged), 0.05, -0.05),
         cont_africa = cont_africa + ox) %>%
  select(cont_africa, everything()) %>%
  
  # plot
  ggplot(aes(x = cont_africa, color = factor(ox))) +
  geom_smooth(data = f,
              aes(y = Estimate, ymin = Q2.5, ymax = Q97.5,
                  fill = factor(ox), linetype = factor(ox)),
              stat = "identity", 
              alpha = 1/4, size = 1/2) +
  geom_point(aes(y = log_gdp),
             alpha = 1/2, shape = 1) +
  scale_colour_pander() +
  scale_fill_pander() +
  scale_x_continuous("Continent", breaks = 0:1, 
                     labels = c("other", "Africa")) +
  coord_cartesian(xlim = c(-.2, 1.2)) +
  ylab("log GDP from year 2000") +
  theme_pander() +
  theme(text = element_text(family = "Times"),
        legend.position = "none")
```

# 连续交互效应

```r
library(rethinking)
data(tulips)
d <- tulips
str(d)
detach(package:rethinking, unload = T)
library(brms)
rm(tulips)
```

- `blooms`：模型的结果变量列一一我们想要预测的量
- `water`：3 个取值分别代表从低（1) 到高（3) 的三种湿度情况
- `shade`：3 个取值分别代表从低（1) 到高（3) 的三种光照情况
- `bed`：代表温室的编号。这些花来自 a、b 和 c 这 3 个温室

## 未中心化的模型
- 无交互：
\begin{aligned} \text { blooms }_{i} & \sim \operatorname{Normal}\left(\mu_{i}, \sigma\right) \\ \mu_{i} &=\alpha+\beta_{1} \text { water }_{i}+\beta_{2} \text { shade }_{i} \\ \alpha & \sim \operatorname{Normal}(0,100) \\ \beta_{1} & \sim \operatorname{Normal}(0,100) \\ \beta_{2} & \sim \operatorname{Normal}(0,100) \\ \sigma & \sim \operatorname{Uniform}(0,100) \end{aligned}
- 有交互：
\begin{aligned} \text { blooms }_{i} & \sim \operatorname{Normal}\left(\mu_{i}, \sigma\right) \\ \mu_{i} &=\alpha+\beta_{1} \text { water }+\beta_{2} \text { shade }_{i}+\beta_{3} \text { water }_{i} \times \text { shade }_{i} \\ \alpha & \sim \operatorname{Normal}(0,100) \\ \beta_{1} & \sim \operatorname{Normal}(0,100) \\ \beta_{2} & \sim \operatorname{Normal}(0,100) \\ \beta_{3} & \sim \operatorname{Normal}(0,100) \\ \sigma & \sim \text { Uniform }(0,100) \end{aligned}

```r
b7.6 <-
  brm(data = d, family = gaussian,
      blooms ~ 1 + water + shade,
      prior = c(prior(normal(0, 100), class = Intercept),
                prior(normal(0, 100), class = b),
                prior(uniform(0, 100), class = sigma)),
      iter = 10000, warmup = 2000, cores = 4, chains = 4,
      seed = 7)

b7.7 <- 
  update(b7.6, 
         formula = blooms ~ 1 + water + shade + water:shade)   # 假设存在交互
```

R 中的模型拟合引擎 optim 花了太长时间试图搜索最优值。其搜索的时间超过了限度，也就是消息中的“最大迭代”（maximum Iterations）。得到的估计不可靠，因此不能使用，除非你能肯定结果是没有问题的。在这里报错，得到的结果不可用。使用`adapt_delta`为σ设置更好的先验

```r
b7.6 <-
  update(b7.6,
         prior = c(prior(normal(0, 100), class = Intercept),
                   prior(normal(0, 100), class = b),
                   prior(cauchy(0, 10), class = sigma)),
         control = list(adapt_delta = 0.9),
         seed = 7)   #进行标准化，此时将对σ设置更好的先验

b7.7 <- 
  update(b7.6, 
         formula = blooms ~ 1 + water + shade + water:shade)
```

```r
posterior_summary(b7.6) %>% round(digits = 2)
posterior_summary(b7.7) %>% round(digits = 2)
```

截距变化太大很难解释

```r
b7.6 <- add_criterion(b7.6, "waic")
b7.7 <- add_criterion(b7.7, "waic")

w <- loo_compare(b7.6, b7.7, criterion = "waic")

print(w, simplify = F)
```

```r
cbind(waic_diff = w[, 1] * -2,
      se        = w[, 2] *  2)

model_weights(b7.6, b7.7, weights = "waic")
```

该比较结果几乎将所有的权重都赋予了 m7.7。因此，让我们考虑 m7.7 的后验分布。现在两个主效应`b_water`和`b_shade`都是正的，但是新的交互效应后验分布`b_water:shade`均值是负数。现在你能说无光照有助于郁金香开花吗？负交互效应本身表明随着无光照处理的增强，水分对花朵的影响减少。但减少了多少呢？

## 中心化并重估计

```r
d <-
  d %>%
  mutate(shade_c = shade - mean(shade),    #类别变量无需用scale
         water_c = water - mean(water))

b7.8 <-
  brm(data = d, family = gaussian,
      blooms ~ 1 + water_c + shade_c,
      prior = c(prior(normal(130, 100), class = Intercept),
                prior(normal(0, 100), class = b),
                prior(cauchy(0, 10), class = sigma)),
      iter = 10000, warmup = 5000, chains = 4, cores = 4,
      control = list(adapt_delta = 0.9),
      seed = 7)

b7.9 <- 
  update(b7.8, 
         formula = blooms ~ 1 + water_c + shade_c + water_c:shade_c)
```

```r
posterior_summary(b7.8) %>% round(digits = 2)
posterior_summary(b7.9) %>% round(digits = 2)
```

```r
tibble(model  = str_c("b7.", 8:9)) %>% 
  mutate(fit  = purrr::map(model, get)) %>% 
  mutate(tidy = purrr::map(fit, broom::tidy)) %>% 
  unnest(tidy) %>% 
  filter(term != "lp__") %>% 
  select(term, estimate, model) %>% 
  spread(key = model, value = estimate) %>% 
  mutate_if(is.double, round, digits = 2)
```

是否中心化结果是相似的，但中心化更加稳定

```r
k <- fixef(b7.7)
k[1] + k[2] * 2 + k[3] * 2 + k[4] * 2 * 2

k <- fixef(b7.9)
k[1] + k[2] * 0 + k[3] * 0 + k[4] * 0 * 0
```

## 变量的顺序对于交互项很重要

```r
plot(conditional_effects(b7.9, 
                      effects = "shade_c:water_c"),
     points = T)

plot(conditional_effects(b7.9, 
                      effects = "water_c:shade_c"),
     points = T)
```

人们可能想要评估交互作用中的第二项的影响——这里是`water_c`——在平均值和平均值+/-1个标准差之外的值。当我们重新生成图7.7的最后一行时，我们表示了基于`water_c`的值-1、0和1的交互作用。在这里，我们可以通过使用`int_conditions`参数来实现这一点。它需要一个list，因此我们将在其中放入我们需要的`water_c`值

```r
ic <- 
  list(water.c = c(-1, 0, 1))

plot(conditional_effects(b7.9, 
                      effects = "shade_c:water_c",
                      int_conditions = ic),
     points = T)
```
